```python
import glob
import pandas as pd
path = r'/kaggle/input/carprice'
filenames = glob.glob(path + "/*.csv")
car = pd.DataFrame()
for f in filenames:
    df = pd.read_csv(f)
    car = car._append(df, ignore_index = True, sort = True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Araç markalarına göre fiyat dağılımı
plt.figure(figsize=(10, 6))
sns.barplot(x='Brand', y='Price', data=df)
plt.xticks(rotation=90)
plt.title('Car Price Distribution by Brand')
plt.xlabel('Car Brand')
plt.ylabel('Price')
plt.tight_layout()
plt.savefig('car_price_distribution.png')
```

```python
# Motor hacmi ile fiyat 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine_Size', y='Price', data=df)
plt.title('Engine Size vs Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.tight_layout()
plt.savefig('horsepower_vs_price.png')
```

```python
# Şanzıman türüne göre fiyatlar
plt.figure(figsize=(10, 6))
sns.boxplot(x='Transmission', y='Price', data=df)
plt.title('Price Distribution by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Price')
plt.tight_layout()
plt.savefig('boxplot_price.png')
```

```python
corr = df[['Engine_Size', 'Mileage', 'Price']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.savefig('heatmap.png')
```
